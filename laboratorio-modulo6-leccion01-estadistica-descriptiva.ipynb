#### Ejercicio 1: 

1.  Dados los siguientes datos: [15, 21, 25, 32, 35, 45, 50, 50, 55, 60]. Crea una funcion que calcule la media, mediana y moda.


```python
def ejercicio1(valores: list[int]) -> dict:
    return {
        'valores': list(valores),
        'media': np.mean(valores),
        'mediana': np.median(valores),
        'moda': stats.mode(valores)[0]
    }


lista_valores = [15, 21, 25, 32, 35, 45, 50, 50, 55, 60]

ejercicio1(lista_valores)
```

2.	Interpretación de la media: ¿Qué sucede si un valor extremo, como 100, se agrega al conjunto de datos anterior? ¿Cómo cambia la media? Explica por qué sucede esto.

```python
lista_valores2 = [15, 21, 25, 32, 35, 45, 50, 50, 55, 60, 100]

ejercicio1(lista_valores2)
```

<h3 style="color: purple">La media se nos altera porque es el resultado de sumar todos los valores y dividirlos por el numero total de elementos que estamos sumando </h3>

4. **Rango y rango intercuartílico**:
   - Calcula el **rango** y el **rango intercuartílico** para ambas columnas. 
   - ¿Qué te dice esto sobre la variabilidad de los datos? Justifica tus conclusiones con representación de graficas. 

```python
rangos = {}
for columna in ['Ventas Realizadas', 'Horas Trabajadas']:
    rango = df_rendimiento[columna].max() - df_rendimiento[columna].min()

    q1 = df_rendimiento[columna].quantile(0.25)
    q3 = df_rendimiento[columna].quantile(0.75)

    rangos[columna] = {'Rango': rango, 'Rango Intercuartilico': q3 - q1}

pd.DataFrame(rangos)
```

<h3 style="color: purple">Nuestros rangos son algo altos pero los rangos intercuartilicos son mas bajo. Esto nos indica que tenemos outlayers pero que el 50% de los datos estan mas o menos concentrados</h3>

5. **Cálculo de la asimetría**:

6. **Cálculo de la curtosis**:
